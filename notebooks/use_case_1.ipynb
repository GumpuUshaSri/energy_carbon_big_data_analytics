# USE CASE 1 - Energy and Carbon dashboard

## 1. Load the data

```python
# Load the saved table into a Spark DataFrame
df_energy = spark.read.table("data.energy_volume.energy_power_data")

# Preview top rows
display(df_energy)
```

## 2. Summary of the data loaded

```python
from pyspark.sql.functions import col

# Cast columns to float for aggregation
df_energy = df_energy.withColumn(
    "energy_kWh", col("energy_kWh").cast("float")
).withColumn(
    "carbon_kg", col("carbon_kg").cast("float")
)

# Summary statistics
display(df_energy.describe())

```

## 3. Total energy consumption by device type

```python
# Energy consumption by device type
display(
    df_energy.groupBy("device_type")
    .avg("energy_kWh")
    .orderBy("avg(energy_kWh)", ascending=False)
)
```

## 4. Carbon footprint by location

```python
# Carbon footprint by location
display(
    df_energy.groupBy("location")
    .sum("carbon_kg")
    .orderBy("sum(carbon_kg)", ascending=False)
)
```

## 4. KPIs(key performance indicators)

```python
query_kpi = """
SELECT 
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh,
    ROUND(SUM(carbon_kg), 2)  AS total_carbon_kg,
    ROUND(AVG(energy_kWh), 3) AS avg_energy_kWh_per_device,
    ROUND(AVG(carbon_kg), 3)  AS avg_carbon_kg_per_device
FROM data.energy_volume.energy_power_data
"""
display(spark.sql(query_kpi))
```

```python
df_kpi = spark.sql("""
SELECT 
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh,
    ROUND(SUM(carbon_kg), 2)  AS total_carbon_kg,
    ROUND(AVG(energy_kWh), 3) AS avg_energy_kWh_per_device,
    ROUND(AVG(carbon_kg), 3)  AS avg_carbon_kg_per_device
FROM data.energy_volume.energy_power_data
""")

df_kpi.write.mode("overwrite").saveAsTable("data.energy_volume.uc1_kpi")

print(" KPIs saved as data.energy_volume.uc1_kpi")
```

## 5. Daily trend

```python
query_daily_trend = """
SELECT 
    DATE(timestamp) AS date,
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh,
    ROUND(SUM(carbon_kg), 2)  AS total_carbon_kg
FROM data.energy_volume.energy_power_data
GROUP BY DATE(timestamp)
ORDER BY date
"""
display(spark.sql(query_daily_trend))
```

## 6. Top device by energy consumption

```python
query_top_devices = """
SELECT 
    device_type,
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh
FROM data.energy_volume.energy_power_data
GROUP BY device_type
ORDER BY total_energy_kWh DESC
LIMIT 5
"""
display(spark.sql(query_top_devices))
```

## 7. Efficiency

```python
query_efficiency = """
SELECT
    device_type,
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh,
    ROUND(SUM(carbon_kg), 2) AS total_carbon_kg,
    ROUND(SUM(energy_kWh) / SUM(carbon_kg), 3) AS energy_per_carbon_ratio
FROM data.energy_volume.energy_power_data
GROUP BY device_type
ORDER BY energy_per_carbon_ratio DESC
"""
display(spark.sql(query_efficiency))
```

```python
# Save summary metrics as a table
df_summary = spark.sql("""
SELECT 
    DATE(timestamp) AS date,
    ROUND(SUM(energy_kWh), 2) AS total_energy_kWh,
    ROUND(SUM(carbon_kg), 2)  AS total_carbon_kg
FROM data.energy_volume.energy_power_data
GROUP BY DATE(timestamp)
ORDER BY date
""")

# Persist output
df_summary.write.mode("overwrite").saveAsTable("data.energy_volume.uc1_summary")

print(" Use Case 1 Summary Saved as data.energy_volume.uc1_summary")
```
