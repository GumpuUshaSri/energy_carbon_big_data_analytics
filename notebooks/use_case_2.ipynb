## 3. Imports & load table

```python
# Imports
from pyspark.sql.functions import col, date_trunc, hour, dayofweek, to_timestamp, avg, sum as _sum, lag
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml.clustering import GaussianMixture
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
from pyspark.sql.types import DoubleType
import pyspark.sql.functions as F

# Load table (existing)
df = spark.read.table("data.energy_volume.energy_power_data")
df = df.withColumn("timestamp", to_timestamp(col("timestamp")))
df.createOrReplaceTempView("energy_data")
display(df.limit(5))
```

## 4. Feature engineering(hourly aggregates & usage patterns)

```python
# Aggregate to hourly per device (helps modeling)
df_hour = df.withColumn("hour_ts", date_trunc("hour", col("timestamp"))) \
            .groupBy("device_id", "device_type", "location", "hour_ts") \
            .agg(
                F.round(F.avg("energy_kWh"), 6).alias("avg_energy_kWh"),
                F.round(F.avg("carbon_kg"), 6).alias("avg_carbon_kg"),
                F.min("timestamp").alias("sample_ts")
            ).withColumnRenamed("hour_ts", "timestamp")

# Add temporal features
df_hour = df_hour.withColumn("hour", hour(col("timestamp"))) \
                 .withColumn("day_of_week", dayofweek(col("timestamp")))

# Add rolling/lag features per device for forecasting
w = Window.partitionBy("device_id").orderBy("timestamp")
df_hour = df_hour.withColumn("lag_1", F.lag("avg_energy_kWh", 1).over(w)) \
                 .withColumn("lag_2", F.lag("avg_energy_kWh", 2).over(w)) \
                 .withColumn("lag_3", F.lag("avg_energy_kWh", 3).over(w))

# Fill nulls in lags with avg energy per device as fallback
device_avg = df_hour.groupBy("device_id").agg(F.mean("avg_energy_kWh").alias("dev_avg"))
df_hour = df_hour.join(device_avg, on="device_id", how="left") \
                 .withColumn("lag_1", F.coalesce(col("lag_1"), col("dev_avg"))) \
                 .withColumn("lag_2", F.coalesce(col("lag_2"), col("dev_avg"))) \
                 .withColumn("lag_3", F.coalesce(col("lag_3"), col("dev_avg"))) \
                 .drop("dev_avg")

# Persist intermediate DF for speed
df_hour.createOrReplaceTempView("df_hour")
display(df_hour.limit(5))
```

## 5. Clustering devices by usage pattern(KMeans)

```python
# Compute mean energy per device, type, and location
device_feats = (
    df.groupBy(
        "device_id",
        "device_type",
        "location"
    )
    .agg(
        F.mean("energy_kwh").alias("mean_energy")
    )
)

# Convert to Pandas (only mean_energy selected)
pdf = device_feats.select(
    "device_id", "device_type", "location", "mean_energy"
).toPandas()

# Start with 1% sample
pdf_sampled = pdf.sample(frac=0.01, random_state=42)

# Increase sample fraction until enough samples for clustering
sample_fractions = [0.01, 0.05, 0.1, 0.2, 0.5, 1.0]
for frac in sample_fractions:
    pdf_sampled = pdf.sample(frac=frac, random_state=42)
    if len(pdf_sampled) >= 3:  # n_clusters=3
        break

if len(pdf_sampled) >= 3:
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters=3, random_state=42)
    pdf_sampled["cluster"] = kmeans.fit_predict(pdf_sampled[["mean_energy"]])

    device_clustered = spark.createDataFrame(pdf_sampled)
    display(device_clustered)

    spark.sql("CREATE SCHEMA IF NOT EXISTS energy_ml")
    device_clustered.write.mode("overwrite").saveAsTable("energy_ml.device_clusters")
else:
    print("Not enough samples for clustering. Please check your data or reduce n_clusters.")
```

## 6. Anomaly detection using GaussianMixture (low-likelihood)

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# 1) Convert to Pandas (only needed feature columns)
pdf = df_hour.select(
    "device_id",
    "timestamp",
    "avg_energy_kWh"
).toPandas()

# 2) Scale
scaler = StandardScaler()
pdf["scaled_energy"] = scaler.fit_transform(pdf[["avg_energy_kWh"]])

# 3) Fit Gaussian Mixture Model (3 components)
gmm = GaussianMixture(
    n_components=3,
    random_state=42
)
pdf["gmm_cluster"] = gmm.fit_predict(pdf[["scaled_energy"]])

# 4) Compute anomaly score = distance to assigned component mean in original scale
means = scaler.inverse_transform(gmm.means_)
pdf["comp_mean"] = pdf["gmm_cluster"].apply(lambda c: means[c][0])
pdf["residual"] = abs(pdf["avg_energy_kWh"] - pdf["comp_mean"])

# 5) Compute anomaly threshold (mean + 3 * std)
r_mean = pdf["residual"].mean()
r_std = pdf["residual"].std() if pdf["residual"].std() != 0 else 0.001
threshold = r_mean + 3 * r_std

# 6) Filter anomalies
pdf_anomalies = pdf[pdf["residual"] > threshold]

# 7) Convert to Spark DataFrame and select columns
df_anomalies = spark.createDataFrame(pdf_anomalies)[
    [
        "device_id",
        "timestamp",
        "avg_energy_kWh",
        "residual"
    ]
]

spark.sql("CREATE SCHEMA IF NOT EXISTS energy_ml")
df_anomalies.write.mode("overwrite").saveAsTable("energy_ml.anomalies")

display(df_anomalies.limit(50))
```

## 7. Short-term forecasting per cluster (lag features + LinearRegression)

```python
# Join device clusters into hour data
device_clustered.createOrReplaceTempView("device_clustered")
df_hour2 = df_hour.join(device_clustered.select("device_id","cluster"), on="device_id", how="left")

# Prepare features: ['lag_1','lag_2','lag_3','hour','day_of_week']
feature_cols = ["lag_1","lag_2","lag_3","hour","day_of_week"]
assembler_forecast = VectorAssembler(inputCols=feature_cols, outputCol="features_f")
# We'll run a LinearRegression per cluster and store models/predictions
predictions_union = None

for cl in [0,1,2]:
    df_cl = df_hour2.filter(col("cluster")==cl).na.drop(subset=feature_cols + ["avg_energy_kWh"])
    if df_cl.count() < 50:
        continue
    lr = LinearRegression(featuresCol="features_f", labelCol="avg_energy_kWh", predictionCol="pred_energy")
    pipeline_cl = Pipeline(stages=[assembler_forecast, lr])
    model_cl = pipeline_cl.fit(df_cl)
    pred = model_cl.transform(df_cl).select("device_id","timestamp","avg_energy_kWh","pred_energy","cluster")
    if predictions_union is None:
        predictions_union = pred
    else:
        predictions_union = predictions_union.union(pred)

# Show predictions
if predictions_union:
    display(predictions_union.limit(50))
    predictions_union.write.mode("overwrite").saveAsTable("energy_ml.cluster_forecasts")
else:
    print("No cluster had enough data for forecasting.")
```

##8. Adaptive recommendations engine (rules + forecast-driven)

```python
# Compute cluster-level thresholds (e.g., 75th percentile)
cluster_stats = df_hour2.groupBy("cluster").agg(
    F.expr("percentile_approx(avg_energy_kWh, 0.75)").alias("p75")
).collect()
cluster_thresholds = {r['cluster']: r['p75'] for r in cluster_stats}

# Read forecasts
forecasts = spark.table("energy_ml.cluster_forecasts")
# join to device info to create recommendations
def action_for_row(pred, cluster):
    thr = cluster_thresholds.get(cluster, 0.1)
    if pred >= thr * 1.3:
        return "Immediate: Force Eco Mode / Shutdown non-critical loads"
    elif pred >= thr:
        return "Schedule: Defer non-critical tasks / Lower AC setpoint"
    else:
        return "Normal"

from pyspark.sql.types import StringType
action_udf = F.udf(action_for_row, StringType())

recs = forecasts.withColumn("action", action_udf(col("pred_energy"), col("cluster")))
display(recs.limit(50))
recs.write.mode("overwrite").saveAsTable("energy_ml.recommendations")
```

## 9. Device co-usage graph (Fallback  Spark SQL alternative to GraphFrames)

```python
# Alternative to GraphFrames for Spark Connect environments
from pyspark.sql import functions as F

# Build co-usage edges using same-hour activity
df_active = df_hour.filter(F.col("avg_energy_kWh") > 0.005).select("timestamp", "device_id")

pairs = df_active.alias("a").join(
    df_active.alias("b"),
    (F.col("a.timestamp") == F.col("b.timestamp")) & (F.col("a.device_id") < F.col("b.device_id")),
    how="inner"
).select(F.col("a.device_id").alias("src"), F.col("b.device_id").alias("dst"))

# Count edge frequency
edges = pairs.groupBy("src", "dst").count().withColumnRenamed("count", "weight")

# Compute simple degree centrality (proxy for influence)
degree_df = edges.groupBy("src").agg(F.sum("weight").alias("degree"))
pagerank_df = degree_df.withColumnRenamed("src", "device_id")

# Normalize scores
total_weight = pagerank_df.agg(F.sum("degree")).collect()[0][0]
pagerank_df = pagerank_df.withColumn("pagerank", F.col("degree") / F.lit(total_weight))

display(pagerank_df.orderBy(F.desc("pagerank")).limit(20))
pagerank_df.write.mode("overwrite").saveAsTable("energy_ml.device_pagerank")
```

## 10. Visualize device co-usage network

```python
# === Visualize Device Co-Usage Network ===

import networkx as nx
import matplotlib.pyplot as plt

# Convert Spark edges to Pandas for visualization
edges_pd = edges.limit(200).toPandas()  # limit to avoid clutter

# Create NetworkX graph
G = nx.Graph()

# Add weighted edges
for _, row in edges_pd.iterrows():
    G.add_edge(row['src'], row['dst'], weight=row['weight'])

# Compute layout and draw
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, k=0.3, seed=42)

# Node size based on influence (degree)
degrees = dict(G.degree())
nx.draw_networkx_nodes(G, pos, node_size=[v * 100 for v in degrees.values()], node_color="skyblue", alpha=0.7)
nx.draw_networkx_edges(G, pos, width=0.5, alpha=0.4)
nx.draw_networkx_labels(G, pos, font_size=8)

plt.title("Device Co-Usage Network (Top 200 Edges)", fontsize=14)
plt.axis("off")
plt.show()
```
